```python
## Scores by School Size

# Establish the bins.

size_bins = [0,1000,2000,5000]
group_names = ["small(<1000)","Medium((1000-2000)", "Large(2000-5000)"]


per_school_summary_df["Size Ranges (Per Student)"] = pd.cut(per_school_counts, size_bins, labels=group_names)
per_school_summary_df

per_school_size_by_scores_avg_math = per_school_summary_df.groupby(["Size Ranges (Per Student)"]).mean()["Average Math Score"]
per_school_size_by_scores_avg_reading = per_school_summary_df.groupby(["Size Ranges (Per Student)"]).mean()["Average Reading Score"]
per_school_size_by_scores_percent_reading = per_school_summary_df.groupby(["Size Ranges (Per Student)"]).mean()["% Passing Reading"]
per_school_size_by_scores_percent_math = per_school_summary_df.groupby(["Size Ranges (Per Student)"]).mean()["% Passing Math"]
per_school_size_by_scores_percent_overall = per_school_summary_df.groupby(["Size Ranges (Per Student)"]).mean()["% Overall Passing"]

size_summary_df = pd.DataFrame({
    "Average Math Score": per_school_size_by_scores_avg_math,
    "Average Reading Score": per_school_size_by_scores_avg_reading,
    "% Passing Reading": per_school_size_by_scores_percent_math,
    "% Passing Math": per_school_size_by_scores_percent_math,
    "% Overall Passing": per_school_size_by_scores_percent_overall})

size_summary_df["Average Math Score"] = size_summary_df["Average Math Score"].map("{:.1f}".format)
size_summary_df["Average Reading Score"] = size_summary_df["Average Reading Score"].map("{:.1f}".format)
size_summary_df["% Passing Reading"] = size_summary_df["% Passing Reading"].map("{:.1f}".format)
size_summary_df["% Passing Math"] = size_summary_df["% Passing Math"].map("{:.1f}".format)
size_summary_df[ "% Overall Passing"] = size_summary_df["% Overall Passing"].map("{:.1f}".format)

#use per school summary school size column

per_school_summary_df



#per_school_summary["School Size"] = pd.cut(per_school_summary_df["Total Students"], size_bins, labels=groupnames])
# Categorize spending based on the bins.


# Calculate averages for the desired columns. 


# Assemble into DataFrame. 


# Format the DataFrame  
```
